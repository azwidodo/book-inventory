```python
import pandas as pd
import json
from find_book import *
import re
```

```python
to_add = set()

with open("books.json", "r+") as f:
    books = json.load(f)

    for book in books["books"]:
        to_add.add(book["ISBN"])

with open("not_yet.txt", "r+") as f:
    for line in f:
        isbn = line.rstrip()
        to_add.add(isbn)
```

```python
data = []

for isbn in to_add:
    title, series, authors = find_book(isbn)

    if title != "None":
        if not series:
            row = [isbn, title, "", authors]
        else:
            ser = re.search(r"\((.*?)\)", series[0]).group(1)
            row = [isbn, title, ser, authors]

        data.append(row)
```

```python
df = pd.DataFrame(data, columns=["ISBN", "title", "series", "authors"])
```

```python
df.to_csv("books.csv", index=False)
```
